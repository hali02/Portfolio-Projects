```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#load data
penguins = sns.load_dataset('penguins')
```

```python
penguins.describe()
```

```python
penguins.isnull().sum()
```

```python
penguins = penguins.dropna()
```

```python
penguins.isnull().sum()
```

```python
penguins.head()
```

```python
# plot 1
ax = sns.regplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
# plot 2 figure-level plot separated by sex
g = sns.lmplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="sex")
```

```python
# plot 3 figure-level plot separated by species
g = sns.lmplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```
